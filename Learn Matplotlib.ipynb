```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
style.use('ggplot')
```

```python
# Draw a Bar Chart of how many people like what kind of movie based on given data
# Reference: https://www.mathsisfun.com/data/bar-graphs.html
# When to use: to plot categorical data

# Type of movies
x = ['Action', 'Comedy', 'Drame', 'Romance', 'SciFi']

# Number of peoples like respective movies
y = [5, 4, 1, 6, 4]

# Plot bar chart based on x and y
plt.bar(x, y, label='Bars', color='c')

# Add labels
plt.xlabel('Movies')
plt.ylabel('People')
plt.title('Favourite Movies')
plt.legend()

# Draw Bar chart
# Note: Sometimes "Bars" are right aligned, not center aligned.
plt.show()
```

```python
# Let's understand Histograms with example
# Reference: https://www.mathsisfun.com/data/histograms.html
# It is used to show distribution of variables such as ages, scores.
# Histograms plot quantitative data with ranges of the data grouped into bins or intervals.
# Each range of bin is termed as class interval.

# Lets measure the change in height(cm) of tree


# Trees with heights in cm
y = [110, 120, 130, 155, 160, 170, 190, 199, 225, 230, 255, 278, 290, 298, 310, 330]

# bins - define range
tree_bins = [100, 150, 200, 250, 300, 350]

# Plot bar chart based on x and y
plt.hist(y, bins=tree_bins, edgecolor='black', color='green')

# Show grids
# plt.grid(True, color='orange')

# Add labels
plt.xlabel('Height(cm)')
plt.ylabel('Trees')
plt.title('Tree Heights')
plt.show()
```
